```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# Create a Model Class that inherits nn.Module
class Model(nn.Module):
  # Input layer (4 featues of the flower) -->
  # H1 or Hidden Layer1(number of neurons -->
  # H2 (n) -->
  # output(3 class of iris flower)

  def __init__(self,in_features=4,h1=8,h2=9,out_features=3):
    super().__init__() # inistantiate out nn.Module
    self.fc1 = nn.Linear(in_features,h1)
    self.fc2 = nn.Linear(h1,h2)
    self.out = nn.Linear(h2,out_features)

  def forward(self,x):
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.out(x)

    return x
```

```python
# Pick a manual seed for randomizations
torch.manual_seed(87)
# Create an instance of model
model = Model()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Iris dataset
url = 'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# Change the last column to integer
df['variety'] = df['variety'].replace('Setosa',0.0)
df['variety'] = df['variety'].replace('Versicolor',1.0)
df['variety'] = df['variety'].replace('Virginica',2.0)
df
```

```python
# Train  Test  Split   Set X, y
X = df.drop('variety', axis=1)
y = df['variety']
```

```python
X
```

```python
y
```

```python
# Convert these to numpy arrays
X = X.values
y = y.values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Train Test  Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=87)
```

```python
# Convert x features to float
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
```

```python
# Convert y labels to tensors long
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
```

```python
# Set the criterion of model to measure the error, how far off the predictions are from the data
criterion = nn.CrossEntropyLoss()
# Choose Adam Optimizer, lr = learning rate (if error does not go down after a bunch of iteration (epochs),lower our learning rate )
optimizer = torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
model.parameters
```

```python
# Tranin our model
# Epoch ( one run thru all the training data in our network)
epochs = 100
losses = []
for i in range(epochs):
  # Go forward and get a prediction
  y_pred = model.forward(X_train) # get predicted results

  # Measure the loss/error
  loss = criterion(y_pred,y_train)

  # Keep Track of out losses
  losses.append(loss.detach().numpy())

  # Print every 10 epoch
  if i % 10 == 0:
    print(f"Epoch: {i} and loss: {loss}")

  # Do some back propagation: take the error rate of forward propagation and feed it back thru the newtwok fo fine tune the weights
  optimizer.zero_grad()
  loss.backward()
  optimizer.step
```

```python
# Graph it out
plt.plot(range(epochs),losses)
plt.ylabel("loss/error")
plt.xlabel('Epoch')
```

```python
# Evaluate Model on Test Date Set (validate model on test set)
with torch.no_grad(): # Basically turn off back propagation
  y_eval = model.forward(X_test) # X_test are features from our test set, y_eval will be predictions
  loss = criterion(y_eval,y_test) # Find the loss
```

```python
loss
```

```python
correct = 0
with torch.no_grad():
  for i, data in enumerate(X_test):
    y_val = model.forward(data)

    # Will tell us what type of flower class our network thinks it is
    print(f'{i+1}. {str(y_val)}\t {y_test[i]} \t {y_val.argmax().item()}')

    # Correct or not
    if y_val.argmax().item() == y_test[i]:
      correct += 1
print(f"We got {correct} correct")
```
